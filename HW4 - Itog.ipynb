## Задача 3
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока. Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.


```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
```

### Имеется выраженная зависимость цены от жилой площади, зависимость более чёткая в нижнем ценовом диапазоне и для небольших площадей.

```python
sns.boxplot(x=df['price'], y=df['condition'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('condition')
plt.title('Распределение цены по состоянию');
```

#### Не смотря на низкий коэффициент корреляции, на диаграмме прослеживается некоторое увеличение цены для домов в лучшем состоянии. Слабая выраженность зависимости связана с малой долей домов в плохом состоянии.

```python
sns.boxplot(x=df['price'], y=df['floors'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('floors')
plt.title('Распределение цены в заивисимости от кол-ва этажей');
```

#### Для домов с этажностью с 1 по 2.5 прослеживается равномерный рост цены. 
#### Изменение закономерности для домов с этажностью 3.5 может быть связана с очень небольшой их долей.

```python
sns.boxplot(x=df['price'], y=df['view'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('view')
plt.title('Распределение цены по оценки вида');
```
